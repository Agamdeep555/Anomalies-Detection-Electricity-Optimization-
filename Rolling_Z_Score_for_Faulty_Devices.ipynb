```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('electricity_appliance_wise_data.csv')
df1['Date'] = pd.to_datetime(df1['Date'])
df1 = df1[df1['Date'].dt.year == 2021]
df1.reset_index(drop=True, inplace=True)
df1.head(2)
```

```python
def zscore(x, window):
    r = x.rolling(window=window)
    m = r.mean().shift(1)
    s = r.std(ddof=0).shift(1)
    z = (x-m)/s
    return z
```

```python
df1['kap_zscore'] = zscore(df1['Kitchen Appliances'], 30)
df1['fridge_zscore'] = zscore(df1['Fridge'], 30)
df1['ac_zscore'] = zscore(df1['AC'], 30)
df1['oap_zscore'] = zscore(df1['Other Appliances'], 30)
df1['wm_zscore'] = zscore(df1['Washing Machine'], 3)
```

```python
import plotly.graph_objects as go

fig = go.Figure()
df_anoms = df1[df1['wm_zscore'] > 3]
fig.add_trace(go.Scatter(x=df1['Date'], y=df1['Washing Machine'],
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Scatter(x=df_anoms['Date'], y=df_anoms['Washing Machine'],
                    mode='markers',
                    name='lines'))
fig.update_layout(title='Anomalies detected in Washing Machine')
```

```python
import datetime
import plotly
import plotly.graph_objects as go

# Detect anomalies in different appliances
df_anoms_fridge = df1[df1['fridge_zscore'] > 3]
df_anoms_kap = df1[df1['kap_zscore'] > 3]
df_anoms_ac = df1[df1['ac_zscore'] > 3]
df_anoms_oap = df1[df1['oap_zscore'] > 3]
df_anoms_wm = df1[df1['wm_zscore'] > 3]

# Generate visualizations for washing machine anomalies
for date in df_anoms_wm['Date']:
    fig = go.Figure()

    # Filter data for specific anomaly date
    df_anoms_temp = df_anoms_wm[df_anoms_wm['Date'] == date]
    df_anoms_temp.reset_index(drop=True, inplace=True)

    # Get historical data up to 1 day after anomaly
    df_all_temp = df1[df1['Date'] <= date + datetime.timedelta(days=1)]
    df_all_temp.reset_index(drop=True, inplace=True)

    # Create visualization
    fig.add_trace(go.Scatter(
        x=df_all_temp['Date'],
        y=df_all_temp['Washing Machine'],
        mode='lines',
        name='Actual Consumption'
    ))

    fig.add_trace(go.Scatter(
        x=df_anoms_temp['Date'],
        y=df_anoms_temp['Washing Machine'],
        mode='markers',
        name='High Fluctuations'
    ))

    fig.update_traces(marker=dict(
        size=5,
        line=dict(width=5, color='red')
    ))

    # Format filename with clean date string
    clean_date = str(date)[:-9]  # Remove time portion
    filename = f'washing_machine_anomaly_{clean_date}.pdf'

    # Save visualization
    plotly.io.write_image(fig, filename, format='pdf')
    print(f'Plot saved successfully for Date: {clean_date}')
```
